#### A review of the .apply() method

```python
bond = pd.read_csv('jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_a_millions(number):
    return str(number) + " MILLIONS!"

#bond['Box Office'] = bond['Box Office'].apply(convert_to_string_and_a_millions)
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_a_millions)
```

```python
bond.head()
```

#### The .apply() method with rows

```python
bond = pd.read_csv('jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return "The Best"
    elif actor == 'Roger Moore' and budget > 40:
        return "Enjoyable"
    else:
        return 'I have no clue'
```

```python
bond.apply(good_movie, axis = 'columns')
```
