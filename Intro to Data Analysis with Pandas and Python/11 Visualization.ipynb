#### Bar graphs

```python
google = data.DataReader(name = 'GOOGL', data_source = 'yahoo', start = '2004-01-01', end = '2006-12-31')
google.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return 'Poor'
    elif stock_price > 200 and stock_price <= 500:
        return 'Satisfactory'
    else:
        return 'Stellar'
```

```python
google['Close'].apply(rank_performance).value_counts()
```

```python
plt.style.use('ggplot')
google['Close'].apply(rank_performance).value_counts().plot(kind = 'bar') #barh
```

#### Pice charts

```python
apple = data.DataReader(name = 'AAPL', data_source = 'yahoo', start = '2012-01-01', end = '2016-12-31')
apple.head(3)
```

```python
apple['Close'].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 23.340429820771817:
        return 'Above average'
    else:
        return 'Below average'
```

```python
plt.style.use('ggplot')
apple['Close'].apply(rank_performance).value_counts().plot(kind='pie', legend = True)
```

#### Histograms

```python
google = data.DataReader(name = 'GOOGL', data_source = 'yahoo', start = '2012-01-01', end = '2016-12-31')
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price/100.0)*100
```

```python
custom_round(113.6)
```

```python
google['High'].apply(custom_round).value_counts().sort_index()
```

```python
google['High'].apply(custom_round).nunique()
```

```python
google['High'].apply(custom_round).plot(kind='hist', bins=9)
```
